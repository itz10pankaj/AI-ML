### Assignment 6: Finding and Replacing Text

Write a function that finds and replaces all occurrences of a given word in a file named `data.txt` with another word.

```python
def find_and_replace(filename,oldWord,newWord):
    with open(filename,'r') as file:
        text=file.read()
    new_text=text.replace(oldWord,newWord)
    with open(filename,'w') as file:
        file.write(new_text)
find_and_replace('example.txt','a','Pankaj')
```

```python
def reverse_line(file_path):
    with open(file_path,'r') as file:
        lines=file.readlines()
    for line in reversed(lines):
        print(line.strip())
```

```python
## Write a function that splits a large file named `large.txt`
#  into smaller files of 100 lines each.
def split_file(fileName):
    with open(fileName,'r') as file:
        file_count=1
        lines=[]
        for line_number,line in enumerate(file,start=1):
            lines.append(line)
            if line_number % 100 == 0:
                with open(f"output_file{file_count}",'w') as out:
                    out.writelines(lines)
                lines=[]
                file_count+=1
        if lines:
            with open(f"output_{file_count}.txt", 'w') as out:
                out.writelines(lines)

split_file("imput.txt")
```

```python
import datetime

def log_message(message, filename='activity.log'):
    timestamp = datetime.datetime.now().isoformat()
    with open(filename, 'a') as file:
        file.write(f'[{timestamp}] {message}\n')

# Test
log_message('This is a log message.')
```

### Assignment 13: CSV File Operations

Write a function that reads a CSV file named `data.csv` and prints its contents as a list of dictionaries.

```python
import csv

def read_csv_as_dicts(filename):
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        return list(reader)
```
