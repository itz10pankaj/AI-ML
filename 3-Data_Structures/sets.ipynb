### Sets

```python
my_set={1,2,3,4,5,67,3,5}
print(type(my_set))
print(my_set)
my_set.add(10)
print(my_set)
my_set.remove(10)  # give error if 10 is not present
print(my_set)
my_set.discard(10) # do not give error is number is not present 
print(my_set)

removed_element=my_set.pop()   # Pop rmeove First Element always
print(removed_element)
print(my_set)

# my_set.clear() #Clear all the elements

print(3 in my_set)
```

### Mathmematical Opt


```python
mySet1={1,2,3,5,6}
mySet2={1,2,4,7,8}

union_Set=mySet1.union(mySet2)
print(union_Set)

intersection_Set=mySet1.intersection(mySet2) # Return New set
print(intersection_Set)

print("set1.........",mySet1)
print("set2.........",mySet2)

mySet1.intersection_update(mySet2) # modify old set

print("set1.........",mySet1)
print("set2.........",mySet2)
```

```python
mySet1={1,2,3,5,6}
mySet2={1,2,4,7,8}

difference1=mySet1.difference(mySet2)
difference2=mySet2.difference(mySet1)
print(difference1)
print(difference2)

mySet1.symmetric_difference(mySet2)

```

```python
## Sets Methods
set1={1,2,3,4,5}
set2={3,4,5}

## is subset
print(set1.issubset(set2))
print(set2.issubset(set1))
print(set1.issuperset(set2))
```

```python
### Counting Unique words in text

text="In this tutorial we are we discussing about sets"
words=text.split()

## convert list of words to set to get unique words
print(words)
print(type(words))
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
