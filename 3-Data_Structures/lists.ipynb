##### List Comprehension

```python
SquareArray = [x**2 for x in range(1,11)]
print(SquareArray)

evenNumbers = [x for x in range(1,11) if x%2==0]
print(evenNumbers)
```

```python
def flatten_list(nested_list):
    flattened_list =[]
    for subloist in nested_list:
        for item in subloist:
            flattened_list.append(item)
    return flattened_list
nested_list = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
flattened = flatten_list(nested_list)
print("Original nested list:")
print(nested_list)
print("Flattened list:")
print(flattened)
```
