### Funtions

#### A funtion is a block of code that performs aa specific task.

```python
## Syntax
def function_name(parameter):
    print(parameter)
    return "2"

print(function_name("Pankaj"))
```

```python
## Default Parameters

def greet(name="Pankaj"):
    print(f"Hello {name} Welcome To the paradise")

greet("Krish")
greet("")
greet()
```

```python
## ### Variable Length Arguments
## Positional And Keywords arguments
## Positional
def print_numbers(*args):
    for num in args:
        print(num)

print_numbers(1,2,3,4,5,6,"krish")

## Keywords
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
print_details(name="krish",age=32,count="2")
```

```python
## We can Also use them as Mix

def print_both(*args,**kargs):
    for arg in args:
        print("args = ",arg)
    for kargI,kargV in kargs.items():
        print("karg = ",kargI,"=",kargV)
print_both(1,2,3,name="Paankaj",age=32)
```
