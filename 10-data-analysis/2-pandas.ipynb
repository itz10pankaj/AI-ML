### Pandas
##### Pandas is a powerful data manipulation library in Python,widly used for data analysis and dala cleaning. It provide two primary data structures: Series and DataFrame. A series is a one dimensioanl array-like object while DataFrame is a two dimesional,size mutabke and potentioally heterogenous tabular data structure with labeled axes

```python
!pip install pandas
import pandas as pd
```

```python
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
## Create a series from a dictonary
data={'a':1,'b':2,'c':3,'d':4}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame
data={
    'Name':['Krish','John','Jack'],
    'Age':[25,26,27],
    'City':['Banglore','New York','Florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a Data frame From a List of Dictionaries

data=[
    {'Name':'Krish','Age':32,'City':'Bangalore'},
    {'Name':'John','Age':34,'City':'Bangalore'},
    {'Name':'Bappy','Age':32,'City':'Bangalore'},
    {'Name':'JAck','Age':32,'City':'Bangalore'}
    
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
