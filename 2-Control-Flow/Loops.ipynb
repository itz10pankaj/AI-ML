# Loops

### For Loop

```python
n = 5
for i in  range (5):
    print("Hello World", i)
```

```python
n =[int(x) for x in input("Enter two numbers").split()]
sum=0
for i in range(min(n[0],n[1]),max(n[0],n[1])+1):
    print(i)
    sum=sum+i
print("Sum is:",sum)
```

## While Loop

```python
count =0 
while(count<5):
    print("Hello World",count)
    count+=1
```

```python
# Loop Control statment 
count=1
while(count>0):
    if count==3:
        break
    print("Hello World",count)
    count+=1
```

```python
# Loop Control statment 
count=1
while(count>0 and count<6):
    if count==3:
        count+=1
        continue
    print("Hello World",count)
    count+=1
```

```python
## Pass --- Do nothing just pass

for i in range(5):
    if i==3:
        print("i is 3 so pass executed",i)
        pass
    print("Hello World",i)
```
