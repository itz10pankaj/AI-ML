#### Generators
Generators are a simpler way to create iterators. They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

```python
## It is subclass of Iterator itself

def square(n):
    for i in range(3):
        yield i**2

print(square(3))
print(type(square(3)))
```

```python
for i in square(3):
    print(i)
```

```python
a=square(3)
a
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
    
```

```python
gen = my_generator()
gen
```

```python
## Generator are particulary useful for reading large fileds because the 
## allow you to prcess one line at a time without loading the entire file 
# into memory 
```

```python
def read_large_file(filepath):
    with open(filepath,'r') as file:
        for line in file:
            yield line
```

```python
file_path='large.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
