## Deep Dive into Operators

```python
## Arithmethic Operation

a=10
b = 3

add_result = a + b
sub_result=a-b
mult_result = a* b
div_result = a/b
floor_div_result = a//b
modulus_result = a % b
exponent_result = a ** b


print("add_result", add_result)
print("sub_result", sub_result)
print("mult_result", mult_result)
print("div_result", div_result)
print("floor_div_result", floor_div_result)
print("modulus_result", modulus_result)
print("exponent_result", exponent_result)
```

```python
## comparsion operators

x = 10
y = 20
print("x == y:", x == y)
print("x != y:", x != y)
print("x > y:", x > y)
print("x < y:", x < y)
print("x >= y:", x >= y)
print("x <= y:", x <= y)
```

```python
## Logical Operators

a = True
b = False
print(a and b)
print(a or b)
print(not a)
print(not b)    
```

```python
## Reverse string 

a = "Pankkaj"
reversed_a = a[::-1]
print("Reversed string:", reversed_a)
```
